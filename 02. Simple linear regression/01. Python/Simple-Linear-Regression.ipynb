# Simple linar regression


## Import the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv("./Salary_Data.csv")
# getting feature variable from first to third column and including all rows (data frame)
feature_x = dataset.iloc[ : , :-1].values
# getting dependent varibale from the very last column (data frame)
dependent_y = dataset.iloc[:, -1].values

# print(feature_x)
# print(dependent_y)
```

## Splitting dataset into training and test set

```python
from sklearn.model_selection import train_test_split
# X = features
# Y = dependent
X = feature_x
y = dependent_y
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

## Training the simple linear regression model on a training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)



# salary_prediction = regressor.predict(X_test)
# years_experience = X_test
# actual_salary = y_test

# plt.figure(figsize=(12, 8))
# plt.plot(years_experience, salary_prediction, label = "Simple linear regression prediction", color = "blue" )
# plt.legend()

# plt.scatter(years_experience, actual_salary, label = "actual data ", color = "red")
# plt.legend()

# plt.tight_layout()
# plt.grid(True)
# plt.show()
```

## Predicting the test result

```python
y_pred = regressor.predict(X_test)
print(regressor.predict(X_test))
```

## visualizing the training set result

```python

plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title("visualizing the training result")
plt.xlabel("years of experience")
plt.ylabel("salary")
plt.grid(True)
plt.show()
```

## visualizing the test result

```python

plt.scatter(X_test, y_test, color='red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.plot(X_test, regressor.predict(X_test), color = 'green')
plt.title("visualizing the test result")
plt.xlabel("years of experience")
plt.ylabel("salary")
plt.grid(True)
plt.show()
```
