# Decision Tree regression

## Importing the libraries

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/06. Decision-tree/Position_Salaries.csv")
X = dataset.iloc[ : , 1:-1 ]
y = dataset.iloc[:, -1]
```

## training the decision tree from the dataset

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X, y)
```

## predicting result

```python
print(regressor.predict([[6.5]]))
```

## visualizing

```python
import matplotlib.ticker as mticker
X_grid = np.arange(min(X['Level']), max(X['Level']), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X['Level'], y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Truth or Bluff (Decision Tree Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.grid(True)
plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x:,.0f}'))
plt.show()
```
